```python
data = pd.read_csv('/kaggle/input/world-university-rankings-2023/World University Rankings 2023.csv')
# Dataframe
print(len(data))
df = pd.DataFrame(data)
df.isna().sum()
```

```python
# Clean data
df = df.dropna() # Drop columns that contains NA value(s)
df.isna().sum()
```

```python
condition = ((df['University Rank'].str.contains("â€“")) | (df['University Rank'] == '1501+'))
row_index = df[condition].index
df.drop(row_index, inplace=True)
df
```

```python
df['University Rank'] = pd.to_numeric(df['University Rank'], errors='coerce')
df.dtypes
```

```python
import matplotlib.pyplot as plt
print(df['University Rank'].dtypes)
plt.scatter(x=df['Teaching Score'], y=df['University Rank'])
plt.xlabel("Teaching Score")
plt.ylabel("Universtiy Rank")
plt.show()
```

We can clearly see a simple linear regression model here. The higher the teaching score, the higher (lower in number) the rank of the university. Now we construct a linear model in scikit learn to demonstrate this situation.

```python
# Split into train and test dataset
from sklearn.model_selection import train_test_split
features = df['Teaching Score'].values.reshape(-1, 1)
targets = df['University Rank'].values.reshape(-1, 1)
features_train, features_test, targets_train, targets_test = train_test_split(features, targets, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(features, targets)
```

```python
# Make prediction on test dataset
targets_pred = model.predict(features_test)
# Find MSE to evaluate the accuracy of the model
from sklearn import metrics
mse = metrics.mean_squared_error(targets_test, targets_pred)
mse
r2 = metrics.r2_score(targets_test, targets_pred)
r2
```
